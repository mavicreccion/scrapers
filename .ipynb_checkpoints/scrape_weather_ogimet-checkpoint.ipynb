```python
from lxml import html
import requests
import pandas as pd
import datetime
from datedelta import datedelta
```

```python
def addMonths(date, numMonths):
    return date + datedelta(months=numMonths)
```

```python
def addDays(date, numDays):
    return date + datedelta(days=numDays)
```

```python
def text(elt):
    return elt.text_content().replace(u'\xa0', u' ')
```

```python
def getText(elt):
    isImg = elt.xpath('//img/@onmouseover')
    if (len(isImg) == 0):
        return text(elt)
    else:
        return isImg[0]
```

```python
URL = 'http://www.ogimet.com/cgi-bin/gsynres?lang=en&ind=98425&decoded=yes&ndays=15&ano=2018&mes=1&day=1&hora=00'
page = requests.get(URL)
root = html.fromstring(page.content)

# retrieve table
table = root.xpath('//table[@bgcolor="#d0d0d0"]')[0]

isImg = table.xpath('//th/img/@onmouseover')

len(isImg)
```

```python
def getData(root):
    # retrieve table
    table = root.xpath('//table[@bgcolor="#d0d0d0"]')[0]

    # get headers
    header = [text(th) for th in table.xpath('//th')]       

    # get data
    data = [[text(td) for td in tr.xpath('td')]  
               for tr in table.xpath('//thead/tr')]

    # merge date and time
    for i in range(1, len(data)):
        datetime_object = datetime.datetime.strptime(data[i][0] + ' ' + data[i][1], '%m/%d/%Y %H:%M')
        data[i] = data[i][2:len(data[i])]
        data[i].insert(0, datetime_object)

    # only add rows with complete data
    data = [row for row in data if len(row)==len(header)]    

    # create dataframe
    data = pd.DataFrame(data, columns=header)  
    
    return data
```

```python
def countRemainingDays(prev_date):
    next_year = prev_date.year + 1
    curr_date = prev_date
    while (curr_date.year < next_year):
        curr_date = curr_date + datedelta(days=1)
    temp_date = curr_date - datedelta(days=1)
    remaining_days = temp_date.day - prev_date.day + 1
    return remaining_days, curr_date
```

```python
curr_date = datetime.date(2015, 1, 1)
end_of_year = False
data_list = []
for i in range(0,24):
    
    # 1. preparing the URL
    remaining_days = 50
    
    if(end_of_year == True):
        days_date = countRemainingDays(curr_date)
        remaining_days = days_date[0]
        curr_date = days_date[1]
        end_of_year = False
    else:
        curr_date = addDays(curr_date, remaining_days) 
    
    URL = 'http://www.ogimet.com/cgi-bin/gsynres?lang=en&ind=98425&decoded=yes&ndays=' + str(remaining_days) + '&ano=' + str(curr_date.year) + '&mes=' + str(curr_date.month) + '&day=' + str(curr_date.day) + '&hora=00'
    print(URL)
    
    # 2. get the html data from website
    page = requests.get(URL)
    root = html.fromstring(page.content)
    
    # 3. get data from html
    data = getData(root)
    
    if(i in [6, 14, 22]):
        i = i - 1
        end_of_year = True
        
    # 4. remove excess first row
    if(i not in [7, 15]):
        data = data.drop(data.index[0])
        
    # 5. reverse data frame order
    data = data.iloc[::-1]
        
    # 6. add to list of df
    data_list.append(data)
```

```python
main_data = pd.concat(data_list, ignore_index=True)
main_data.to_csv('weather_ogimet_2015_2016_2017.csv')
main_data
```
