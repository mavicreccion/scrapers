```python
from lxml import html
import requests
import pandas as pd
from datetime import datetime
```

```python
page = requests.get('http://www.ogimet.com/cgi-bin/gsynres?lang=en&ind=98425&decoded=yes&ndays=20&ano=2015&mes=01&day=21&hora=00')
root = html.fromstring(page.content)
```

```python
elem = root.xpath('//table[@bgcolor="#d0d0d0"]/thead/tr[2]/td[1]/a/text()')
print(elem)
```

```python
def text(elt):
    return elt.text_content().replace(u'\xa0', u' ')
```

```python
print(root.xpath('//table[@bgcolor="#d0d0d0"]/thead/tr[2]/td[3]/font/text()'))
```

```python
# retrieve table
table = root.xpath('//table[@bgcolor="#d0d0d0"]')[0]

# get headers
header = [text(th) for th in table.xpath('//th')]       

# get data
data = [[text(td) for td in tr.xpath('td')]  
           for tr in table.xpath('//thead/tr')]

# merge date and time
for i in range(1, len(data)):
    datetime_object = datetime.strptime(data[i][0] + ' ' + data[i][1], '%m/%d/%Y %H:%M')
    data[i] = data[i][2:len(data[i])]
    data[i].insert(0, datetime_object)

# only add rows with complete data
data = [row for row in data if len(row)==len(header)]    

# create dataframe
data = pd.DataFrame(data, columns=header)                
    
print(data)
```
